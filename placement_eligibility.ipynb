```python
!pip install faker
from faker import Faker
import pandas as pd
fake = Faker()
num_students = 30

students = []
for student_id in range(1, num_students + 1):
    students.append({
        "student_id": student_id,
        "name": fake.name(),
        "age": fake.random_int(18,25),
        "gender": fake.random_element(elements=("Male", "Female")),
        "email": fake.email(),
        "phone": fake.phone_number(),
        "enrollment_year": fake.random_int(2018, 2023),
        "course_batch": fake.word(),
        "city": fake.city(),
        "graduation_year": fake.random_int(2022, 2027)
    })

students_df = pd.DataFrame(students)

programming = []
for student_id in range(1, num_students + 1):
    programming.append({
        "programming_id": student_id,
        "student_id": student_id,
        "language": fake.random_element(elements=("Python", "C", "SQL", "Java", "C++")),
        "problems_solved": fake.random_int(50, 250),
        "assessments_completed": fake.random_int(10, 20),
        "mini_projects": fake.random_int(5, 10),
        "certifications_earned": fake.random_int(1, 5),
        "latest_project_score": fake.random_int(50, 100)
    })

programming_df = pd.DataFrame(programming)

soft_skills = []
for student_id in range(1, num_students + 1):
    soft_skills.append({
        "soft_skill_id": student_id,
        "student_id": student_id,
        "communication": fake.random_int(50, 100),
        "teamwork": fake.random_int(50, 100),
        "presentation": fake.random_int(50, 100),
        "leadership": fake.random_int(50, 100),
        "critical_thinking": fake.random_int(50, 100),
        "interpersonal_skills": fake.random_int(50, 100)
    })

soft_skills_df = pd.DataFrame(soft_skills)

placements = []
for student_id in range(1, num_students + 1):
    placements.append({
        "placement_id": student_id,
        "student_id": student_id,
        "mock_interview_score": fake.random_int(50, 100),
        "internships_completed": fake.random_int(0, 5),
        "placement_status": fake.random_element(elements=("Ready", "Not Ready", "Placed")),
        "company_name": fake.company(),
        "placement_package": fake.random_int(10000, 100000),
        "interview_rounds_cleared": fake.random_int(1, 5),
        "placement_date": fake.date_this_year()
    })
placements_df = pd.DataFrame(placements)
```

```python
students_df.to_csv('students_df.csv', index=False)
programming_df.to_csv('programming_df.csv',index=False)
soft_skills_df.to_csv('soft_skills_df.csv',index=False)
placements_df.to_csv('placements_df.csv',index=False)
```

```python
conn = sqlite3.connect('placement_eligibility.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER,
    gender TEXT,
    email TEXT,
    phone TEXT,
    enrollment_year INTEGER,
    course_batch TEXT,
    city TEXT,
    graduation_year INTEGER
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS programming (
    programming_id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_id INTEGER,
    language TEXT,
    problems_solved INTEGER,
    assessments_completed INTEGER,
    mini_projects INTEGER,
    certifications_earned INTEGER,
    latest_project_score INTEGER,
    FOREIGN KEY (student_id) REFERENCES students(student_id)
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS soft_skills (
    soft_skill_id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_id INTEGER,
    communication INTEGER,
    teamwork INTEGER,
    presentation INTEGER,
    leadership INTEGER,
    critical_thinking INTEGER,
    interpersonal_skills INTEGER,
    FOREIGN KEY (student_id) REFERENCES students(student_id)
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS placements (
    placement_id INTEGER PRIMARY KEY AUTOINCREMENT,
    student_id INTEGER,
    mock_interview_score INTEGER,
    internships_completed INTEGER,
    placement_status TEXT,
    company_name TEXT,
    placement_package REAL,
    interview_rounds_cleared INTEGER,
    placement_date TEXT,
    FOREIGN KEY (student_id) REFERENCES students(student_id)
);
''')

conn.commit()
```

```python
%%writefile app.py
import pandas as pd
import streamlit as st
import sqlite3


# Connect to SQLite database
conn = sqlite3.connect('placement_eligibility.db')
st.title("Placement Eligibility Checker")

# Sidebar for input criteria
st.sidebar.header("Set Eligibility Criteria")

# User-defined criteria for filtering
min_problems = st.sidebar.slider("Minimum Problems Solved", 0, 300, 50)
min_soft_skills = st.sidebar.slider("Minimum Average Soft Skills Score", 0, 100, 75)
min_assessments = st.sidebar.slider("Minimum Assessments Completed", 0, 20, 10)
min_mini_projects = st.sidebar.slider("Minimum Mini Projects", 0, 10, 3)
min_certifications = st.sidebar.slider("Minimum Certifications Earned", 0, 5, 2)


# SQL Query using user-defined criteria
query = f"""
SELECT name, email, age, gender, city
FROM students
WHERE student_id IN (
    SELECT student_id
    FROM programming
    WHERE problems_solved >= {min_problems}
)
AND student_id IN (
    SELECT student_id
    FROM soft_skills
    WHERE
      (communication + teamwork + presentation + leadership + critical_thinking + interpersonal_skills) / 6 >= {min_soft_skills}
)
AND student_id IN (
    SELECT student_id
    FROM programming
    WHERE assessments_completed >= {min_assessments}
)
AND student_id IN (
    SELECT student_id
    FROM programming
    WHERE mini_projects >= {min_mini_projects}
)
AND student_id IN (
    SELECT student_id
    FROM programming
    WHERE certifications_earned >= {min_certifications}
)

"""

df = pd.read_sql_query(query, conn)

# Display results
st.subheader("Eligible Students")
st.dataframe(df)

query_selection = st.selectbox(
    "Select a query to run:",
    options=[
        "1.Average Programming Performance Per Batch",
        "2.students name starting with A",
        "3.Students With 1 Internship Experience",
        "4.Placement ready students by mock interview score",
        "5.Low Performance Students",
        "6.Students with Communication Score>=70",
        "7.Placement Status Summary",
        "8.Display highest and lowest package values",
        "9.Number of Students Skilled in Python",
        "10.Total number of students placed in each company",
    ]
)
if query_selection == "1.Average Programming Performance Per Batch":
    query = """
    SELECT
      students.course_batch,
      AVG(programming.problems_solved) AS avg_problems_solved,
      AVG(programming.assessments_completed) AS avg_assessments_completed
    FROM students
    JOIN programming ON students.student_id = programming.student_id
    GROUP BY students.course_batch;
    """
    avg_batch = pd.read_sql(query, conn)
    st.title("Placement Analytics")
    st.subheader("Average Programming Performance Per Batch")
    st.dataframe(avg_batch)

elif query_selection == "2.students name starting with A":
    query = """
    SELECT name FROM students WHERE name LIKE 'A%';
    """
    name_A = pd.read_sql(query, conn)
    st.title("NAME")
    st.subheader("Students Name Starting with A")
    st.dataframe(name_A)

elif query_selection == "3.Students With 1 Internship Experience":
    query = """
    SELECT name
    FROM students
    JOIN placements ON students.student_id = placements.student_id
    WHERE internships_completed = 1;
    """
    df = pd.read_sql(query, conn)
    st.subheader("Students With 1 Internship Experience")
    st.metric("Total Students", len(df))
    st.dataframe(df)

elif query_selection == "4.Placement ready students by mock interview score":
    top_n = st.selectbox("Select Top N Students", [3, 5], index=0)


    query = f"""
    SELECT S.name,Pl.mock_interview_score
    FROM Students S
    JOIN Placements Pl ON S.student_id = Pl.student_id
    WHERE Pl.placement_status = "Ready"
    ORDER BY Pl.mock_interview_score DESC
    LIMIT {top_n};
    """
    df = pd.read_sql(query, conn)

    if df.empty:
        st.error("No placement-ready students found.")
    else:
        df.columns = df.columns.str.strip()


        st.markdown(f"## Top {top_n} Placement-Ready Students by Mock Interview Score")
        st.dataframe(df)

        st.bar_chart(df.set_index("name")["mock_interview_score"])

elif query_selection == "5.Low Performance Students":
    st.title("Low Performance Students")
    st.markdown("Click the button below to see students who have solved less than 20 problems and have a communication score below 50.")
    if st.button("Show Low Performance Students"):
        query = """
        SELECT S.name, P.problems_solved, SS.communication
        FROM Students S
        JOIN Programming P ON S.student_id = P.student_id
        JOIN soft_skills SS ON S.student_id = SS.student_id
        WHERE P.problems_solved < 20 AND SS.communication < 50;
        """
        df = pd.read_sql_query(query, conn)
        if not df.empty:
            st.dataframe(df)
        else:
            st.write("No students found with the specified criteria.")

elif query_selection == "6.Students with Communication Score>=70":
    query = """
    SELECT S.name, SS.communication
    FROM Students S
    JOIN soft_skills SS ON S.student_id = SS.student_id
    WHERE SS.communication >= 70;
    """
    df = pd.read_sql_query(query, conn)
    st.subheader("Students with Communication Score >= 70")
    st.dataframe(df)

elif query_selection == "7.Placement Status Summary":
    query = """
    SELECT placement_status, COUNT(*) AS total_students
    FROM Placements
    GROUP BY placement_status;
    """
    df = pd.read_sql_query(query, conn)
    st.subheader("Placement Status Summary")
    st.dataframe(df)

elif query_selection == "8.Display highest and lowest package values":
    query = """
    SELECT 'Highest Package' AS label, MAX(placement_package) AS value FROM Placements
    UNION
    SELECT 'Lowest Package', MIN(placement_package) FROM Placements;
    """
    df = pd.read_sql(query, conn)
    package_choice = st.selectbox("Select Package Type", options=df['label'].unique())
    selected_package = df[df['label'] == package_choice]
    st.write(f"### {package_choice} Information")
    st.write(selected_package)

elif query_selection == "9.Number of Students Skilled in Python":
    query = """
    SELECT COUNT(*) AS python_devs
    FROM Programming
    WHERE language = 'Python' AND certifications_earned > 0;
    """
    df = pd.read_sql(query, conn)
    st.subheader("Students Skilled in Python")
    st.dataframe(df)

elif query_selection == "10.Total number of students placed in each company":
    query = """
    SELECT company_name, COUNT(*) AS placed_students
    FROM Placements
    WHERE placement_status = 'Placed'
    GROUP BY company_name;
    """
    df = pd.read_sql(query, conn)
    selected_company = st.selectbox("Select a company to view placement count", df['company_name'])
    filtered_df = df[df['company_name'] == selected_company]
    st.subheader(f"Placement Count for {selected_company}")
    st.write(f"**Placed Students:** {filtered_df['placed_students'].values[0]}")
```
